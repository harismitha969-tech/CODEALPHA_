```python

from googletrans import Translator
from gtts import gTTS
import IPython.display as ipd
translator = Translator()
source_lang = input("Enter source language code (e.g., 'en' for English): ")
target_lang = input("Enter target language code (e.g., 'ta' for Tamil): ")
text_to_translate = input("Enter text to translate: ")
translation = translator.translate(text_to_translate, src=source_lang, dest=target_lang)
print(f"\n✅ Translated Text: {translation.text}")
tts_choice = input("Do you want to hear the translation? (y/n): ").lower()
if tts_choice == 'y':
    tts = gTTS(translation.text, lang=target_lang)
    tts.save("translated_audio.mp3")
    ipd.display(ipd.Audio("translated_audio.mp3"))
```
